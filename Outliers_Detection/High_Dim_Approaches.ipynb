# High Dimensional Approaches

```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from collections import Counter
from collections import defaultdict
from sklearn.preprocessing import StandardScaler
```

```python
#df = pd.read_csv('../dataset/cmi_internet_imputed.csv')
df = pd.read_csv('../dataset/cmi_internet_imputed.csv')
df
```

```python
X = df.drop(columns=['sii'])
y = df['sii']

num_cols = X.select_dtypes(include=np.number).columns.tolist()
print("Numerical variables used:")
print(num_cols)
print("Shape:", X.shape)
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)

X = pd.DataFrame(
    X,
    columns=num_cols,
    index=df.index
)

X
```

## ABOD

```python
from pyod.models.abod import ABOD
```

```python
clf = ABOD(contamination=0.1, n_neighbors=20)
clf.fit(X)
```

```python
outliers = clf.predict(X) # 0 for inliers, 1 for outliers
np.unique(outliers, return_counts=True)

n_out = abs(outliers[outliers == 1].sum())
perc_out = n_out / len(outliers) * 100

print("Number of ABOD outliers:", n_out)
print("Percentage of ABOD outliers:", round(perc_out, 3), "%")
```

```python
from standardize_labels import standardize_labels 
# 1. import matrix
od_results = pd.read_csv('./Outliers/outliers_matrix.csv').set_index('row_id')

# 2. Unify labels (ABOD uses 1 for outliers, we want 1)
new_standard_labels = standardize_labels(outliers, method_type='pyod_or_manual')

# 3. Add the new column to the existing DataFrame
od_results['outlier_ABOD'] = new_standard_labels

# 4. Overwrite the master file with the added column
od_results.to_csv('./Outliers/outliers_matrix.csv', index_label='row_id')
```

```python
plt.hist(clf.decision_scores_, bins=20)
plt.axvline(np.min(clf.decision_scores_[np.where(outliers==1)]), c='k')
plt.show()
```

```python
# PCA Visualization of outliers
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'], index=df.index)
df_pca = df_pca.join(pd.Series(outliers, name='outlier'))
df_pca
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_pca,
    x='PC1',
    y='PC2',
    hue='outlier',
    palette={0: 'blue', 1: 'red'},
    alpha=0.7,
    edgecolor='none'
)
plt.title('ABOD Outliers in PCA Space')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Outlier')
plt.show()
```
