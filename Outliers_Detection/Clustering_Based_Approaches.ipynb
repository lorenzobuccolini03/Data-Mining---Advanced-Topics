# Clustering-based approaches

```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from collections import Counter
from collections import defaultdict
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../dataset/cmi_internet_imputed.csv')
df
```

```python
X = df.drop(columns=['sii'])
y = df['sii']

num_cols = X.select_dtypes(include=np.number).columns.tolist()
print("Numerical variables used:")
print(num_cols)
print("Shape:", X.shape)
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)

X = pd.DataFrame(
    X,
    columns=num_cols,
    index=df.index
)

X
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
from sklearn.neighbors import NearestNeighbors

def plot_elbow_comparison(X, min_samples_list=[4, 16, 32, 64]):
    plt.figure(figsize=(10, 6))
    
    for ms in min_samples_list:
        # Chiediamo ms vicini
        neighbors = NearestNeighbors(n_neighbors=ms)
        neighbors_fit = neighbors.fit(X)
        distances, _ = neighbors_fit.kneighbors(X)
        
        # Prendiamo l'ultima colonna (indice ms-1)
        # che rappresenta la distanza dal k-esimo vicino
        k_distances = np.sort(distances[:, ms - 1], axis=0)
        
        plt.plot(k_distances, label=f'min_samples={ms}')
    
    plt.xlabel('Punti ordinati per distanza')
    plt.ylabel('Epsilon (distanza dal k-esimo vicino)')
    plt.title('Elbow Method per DBSCAN')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Esempio di chiamata:
plot_elbow_comparison(X)
```

```python
dbscan = DBSCAN(eps=4, min_samples=32) 
dbscan.fit(X)
```

```python
np.unique(dbscan.labels_, return_counts=True)
```

```python
df.loc[np.where(dbscan.labels_==-1)]
```

```python
outliers = dbscan.labels_

n_out = abs(outliers[outliers == -1].sum())
perc_out = n_out / len(outliers) * 100

print("Number of DBSCAN outliers:", n_out)
print("Percentage of DBSCAN outliers:", round(perc_out, 3), "%")
```

```python
from standardize_labels import standardize_labels 
# 1. import matrix
od_results = pd.read_csv('./Outliers/outliers_matrix.csv').set_index('row_id')

# 2. Unify labels (DSCAN uses -1 for outliers, we want 1)
new_standard_labels = standardize_labels(outliers, method_type='cluster')

# 3. Add the new column to the existing DataFrame
od_results['outlier_DBSCAN'] = new_standard_labels

# 4. Overwrite the master file with the added column
od_results.to_csv('./Outliers/outliers_matrix.csv', index_label='row_id')
```

```python
# PCA Visualization of outliers
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'], index=df.index)
df_pca = df_pca.join(pd.Series(outliers, name='outlier'))
df_pca
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_pca,
    x='PC1',
    y='PC2',
    hue='outlier',
    palette={0: 'blue', -1: 'red'},
    alpha=0.7,
    edgecolor='none'
)
plt.title('DBSCAN Outliers in PCA Space')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Outlier')
plt.show()
```

## OPTICS

```python
from sklearn.cluster import OPTICS

# Model fitting
optics = OPTICS(min_samples=32, xi=0.05, min_cluster_size=0.05)
optics.fit(X)

# reachability plot
reachability = optics.reachability_[optics.ordering_]
labels = optics.labels_[optics.ordering_]

plt.figure(figsize=(10, 6))
# I "valloni" nel grafico rappresentano i cluster (punti vicini/densi)
# I "picchi" rappresentano i confini tra cluster o punti isolati
plt.plot(reachability, color='teal', alpha=0.7)
plt.title('OPTICS Reachability Plot')
plt.xlabel('Punti ordinati')
plt.ylabel('Reachability Distance')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# choose eps to identify outliers
eps = 4
optics = OPTICS(min_samples=32, xi=0.05, min_cluster_size=0.05, max_eps=eps)
optics.fit(X)
```

## Clustering Based Local Outlier Factor (CBLOF)

```python
from pyod.models.cblof import CBLOF
```

```python
clf = CBLOF()
clf.fit(X)
```

```python
outliers = clf.predict(X)
np.unique(outliers, return_counts=True)
```

```python
# plot decision scores distribution
plt.hist(clf.decision_scores_, bins=20)  
# plot threshold line where points are classified as outliers (1) vs inliers (0)
plt.axvline(np.min(clf.decision_scores_[np.where(outliers==1)]), c='k')
plt.show()
```

```python
n_out = abs(outliers[outliers == 1].sum())
perc_out = n_out / len(outliers) * 100

print("Number of CBLOF outliers:", n_out)
print("Percentage of CBLOF outliers:", round(perc_out, 3), "%")
```
