# Depth Based Approaches

```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from collections import Counter
from collections import defaultdict
from sklearn.preprocessing import StandardScaler
```

```python
#df = pd.read_csv('../dataset/cmi_internet_imputed.csv')
df = pd.read_csv('cmi_internet_imputed.csv')
df
```

```python
X = df.drop(columns=['sii'])
y = df['sii']

num_cols = X.select_dtypes(include=np.number).columns.tolist()
print("Numerical variables used:")
print(num_cols)
print("Shape:", X.shape)
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)

X = pd.DataFrame(
    X,
    columns=num_cols,
    index=df.index
)

X
```

## EllEnv

```python
from sklearn.covariance import EllipticEnvelope
```

```python
ellenv = EllipticEnvelope(random_state=0)
ellenv.fit(X)
```

```python
outliers = ellenv.predict(X)
```

```python
# predict returns -1 for an outlier and 1 for an inlier
np.unique(outliers, return_counts=True)
```

```python
n_out = abs(outliers[outliers == -1].sum())
perc_out = n_out / len(outliers) * 100

print("Number of EllEnv outliers:", n_out)
print("Percentage of EllEnv outliers:", round(perc_out, 3), "%")
```

```python
plt.hist(ellenv.decision_scores_, bins=20)
plt.axvline(np.min(ellenv.decision_scores_[np.where(outliers==-1)]), c='k')
plt.show()
```

```python
# PCA Visualization of outliers
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'], index=df.index)
df_pca = df_pca.join(pd.Series(outliers, name='outlier'))
df_pca
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_pca,
    x='PC1',
    y='PC2',
    hue='outlier',
    palette={1: 'blue', -1: 'red'},
    alpha=0.7,
    edgecolor='none'
)
plt.title('kNN Outliers in PCA Space')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Outlier')
plt.show()
```
